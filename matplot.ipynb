## line plating


```python
day=[1,2,3,4,5,6,7]
temp=[38,34,23,24,22,41,33]
plt.xlabel('Day')
plt.ylabel('tempreture')
plt.plot(day,temp,linestyle='--',color='gray',marker='o')
plt.grid()
plt.show()
```

## multi line plating

```python
days=[1,2,3,4,5,6,7]
max_t=[55,59,63,73,53,69,65]
min_t=[23,25,32,39,29,40,25]
avg_t=[41,43,49,52,39,48,50]
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')
plt.legend()
plt.grid()
plt.show()
```

## bar ploting

```python
company=['RElience','Tata','Adani']
revenue=[110,90,100]
xpos=np.arange(len(company))
plt.bar(xpos,revenue,color='brown')
plt.show()
```

```python
plt.xticks(xpos,company)
plt.ylabel('Revenue')
plt.title('Indian companies')
plt.bar(xpos,revenue,width=0.2,color='brown')
plt.show()
```

## pie chart

```python
x =['usa','china','japan','britan']
gold =[39,38,27,22]
silver=[41,32,14,21]
bronze=[33,18,17,22]
plt.xlabel=('countries')
plt.ylabel=('medal')
plt.bar(x,gold,0.4,label='gold')
plt.bar(x,silver,0.4,label='silver')
plt.bar(x,bronze,0.4,label='bronze')
plt.legend()
plt.show()
```

```python
x =['usa','china','japan','britan']
gold =[39,38,27,22]
silver=[41,32,14,21]
bronze=[33,18,17,22]
brz_bs=list(np.add(gold,silver))
plt.xlabel=('Countries')
plt.ylabel=('Medal')
plt.bar(x,gold,0.4,label='gold')
plt.bar(x,silver,0.4,bottom=gold,label='silver')
plt.bar(x,bronze,0.4,bottom=brz_bs,label='bronze')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
co=['Apple','Oppo','Vivo','Oneplus','Redmi']
share=[10.85,7,6,10,15]
plt.pie(x=share,labels=co,radius=1,autopct='%.2f%%')
plt.show()
```

```python
co=['Apple','Oppo','Vivo','Oneplus','Redmi']
share=[10.85,7,6,10,15]
plt.title('Share values')
plt.pie(x=share,labels=co,radius=1,autopct='%.2f%%',shadow=True,startangle=90)
plt.show()
```

```python
co=['Apple','Oppo','Vivo','Oneplus','Redmi']
share=[10.85,7,6,10,15]
explode=[0,0,0.2,0,0]
plt.title('Share values')
plt.pie(x=share,labels=co,radius=1,explode=explode,autopct='%.2f%%',shadow=True,startangle=90)
plt.show()
```
